```python
haticenin_verisi = pd.read_csv("/Users/haticecakir/Downloads/Veri Bilimi Dosyalar/il_cinsiyet_ort_evlenme_yasi.csv")
haticenin_verisi
```

```python
import matplotlib .pyplot as plt
plt.scatter(haticenin_verisi["Kadın"], haticenin_verisi["Erkek"], color = "purple")
plt.xlabel("Kadın")
plt.ylabel("Erkek")
plt.title ("İllere Göre Kadın-Erkek Ortalama Evlilik Yaşı Grafiği")
plt.show()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=5)
kmeans.fit(haticenin_verisi[['Kadın','Erkek']])
haticenin_verisi['Küme'] = kmeans.predict(haticenin_verisi[['Kadın','Erkek']])
haticenin_verisi
```

```python
plt.scatter(haticenin_verisi["Kadın"], haticenin_verisi["Erkek"], c=haticenin_verisi ['Küme'])
plt.xlabel("Kadın")
plt.ylabel("Erkek")
plt.title ("İllere Göre Kadın-Erkek Ortalama Evlilik Yaşı Grafiği")
plt.show()
```

```python
plt.scatter(haticenin_verisi["Kadın"], haticenin_verisi["Erkek"], c=haticenin_verisi ['Küme'])
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], c='red')
plt.xlabel("Kadın")
plt.ylabel("Erkek")
plt.title ("İllere Göre Kadın-Erkek Ortalama Evlilik Yaşı Grafiği")
plt.show()
```
